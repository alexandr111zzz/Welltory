В аптеки города N поступил экспресс-тест на COVID-19, у которого вероятность ошибки (как позитивной, так и негативной) равна 0.05. По официальной статистике, прямо сейчас в городе N всего 1% жителей болен ковидом. Один из жителей, не имея симптомов, просто на всякий случай решил сдать этот тест, и внезапно тест показал положительный результат. Оцените вероятность того, что этот житель действительно болен ковидом.

Человек болен

* Вероятность ошибки, что человек здоров = 0.05
* Вероятность правильного результата = 0.95

Человек здоров

* Вероятность ошибки, что человек болен = 0.05
* Вероятность правильного результата = 0.95

Всего больных людей

* 0.01 от общей численности

Рассмотрим случай, человек здоров, результат положительный

Воспользуемся формулой байеса

P(здоров/тест+) = P(здоров) * P(Тест+/Здоров) / P(Тест+), где

* P(здоров) - вероятность того, что человек здоров (1 минус процент больных людей от общей численности)
* Р(Тест+/Здоров) - вероятность того, что у здорового человека оказался положительный тест
* Р(Тест+) - вероятность получить положительный тест (вероятность того, что положительный тест получил больной человек + вероятность того, что положительный тест получил здоровый человек)

```python
# Подставим в формулу исходные данные

p_healthy_test_plus = (0.99 * 0.05) / (0.01 * 0.95 + 0.99 * 0.05)
```

Рассмотрим случай, человек болен, но у него нет симптомов, а результат положительный

P(болен/тест+) = P(болен) * P(Тест+/болен) / P(Тест+), где

* P(болен) - вероятность того, что человек болен
* Р(Тест+/Здоров) - вероятность того, что у больного человека оказался положительный тест
* Р(Тест+) - вероятность получить положительный тест (вероятность того, что положительный тест получил больной человек + вероятность того, что положительный тест получил здоровый человек)

```python
# Подставим в формулу исходные данные

p_ill_test_plus = (0.01 * 0.95) / (0.01 * 0.95 + 0.99 * 0.05)
```

```python
# Итог
print(f'''Вероятность того, что человек здоров, но получил положительный тест равна {round(p_healthy_test_plus, 3)} \n
Вероятность того, что человек болен и получил положительный тест равна {round(p_ill_test_plus, 3)}''')
```
